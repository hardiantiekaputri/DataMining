```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

# Menampilkan dataset breast cancer
data = load_breast_cancer()
X = data.data

# Mencari SSE untuk tiap nilai k
sse = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# Menampilkan plot elbow
sns.set(style='whitegrid')
plt.plot(range(1,11), sse)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Klaster')
plt.ylabel('SSE')
plt.show()
```

```python
# Melakukan Clustering K-Means, untuk n=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# clustering KMeans dengan 2 klaster
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

# menambahkan kolom label pada dataset
labels = kmeans.labels_
df = pd.DataFrame(X)
df['label'] = labels

# membuat scatterplot dengan warna berdasarkan label
sns.scatterplot(x=0, y=1, hue='label', data=df)
plt.show()
```
